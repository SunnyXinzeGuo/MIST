```python
K = 100
sigma = 1.2
```

```python
def Acos(R, theta, miu, nu):
    part1 = K/sigma*np.cos(miu*theta)*(R/sigma)**miu
    part2 = P(miu,nu,R**2/sigma**2)
    part3 = np.exp(-R**2/4/sigma**2)
    return part1*part2*part3
```

```python
def Asin(R, theta, miu, nu):
    part1 = K/sigma*np.sin(miu*theta)*(R/sigma)**miu
    part2 = P(miu,nu,R**2/sigma**2)
    part3 = np.exp(-R**2/4/sigma**2)
    return part1*part2*part3
```

```python
def P(miu,nu,u):
    returnarray = np.ones([360,360])
    for i in range(360):
        for j in range(360):
            terms = []
            p = 0
            while p <= nu:
                up = np.math.factorial(miu + nu)*np.math.factorial(nu)
                down = np.math.factorial(miu + p)*np.math.factorial(p)*np.math.factorial(nu-p)
                term = (-2)**(nu-p)* up/down * u[i][j]**p
                terms.append(term)
                p += 1
            returnarray[i][j] = np.sum(terms)
    return returnarray
```

```python
# new ways of drawing
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.patches as mpatches
from matplotlib.collections import PatchCollection
import pandas as pd

r = xx.ravel()
theta = yy.ravel()
value = zz00c.ravel()
df = pd.DataFrame(list(zip(r,theta,value)),columns =['r','theta','intensity'])
colors = ['#000052','#0c44ac','#faf0ca','#ed0101','#970005'] 
cm = LinearSegmentedColormap.from_list('custom', colors,N=10)
cm.set_bad(color='white')
```

```python
patches = []; avg = []
ntheta = 100
dtheta = 360/ntheta;
nradius = 100 
dradius = 10/nradius;

for nr in range(nradius,0,-1):  # See!  Outside to Inside
    start_r = (nr-1)*dradius
    end_r = (nr)*dradius
    for nt in range(0,ntheta):
        start_t = nt*dtheta
        end_t = (nt+1)*dtheta
    
        stripped = df[(df['r']>=start_r) & (df['r']<end_r) &          
            (df['theta']>=start_t) & (df['theta']<end_t)]
       
        avg.append(stripped['intensity'].mean())
        wedge = mpatches.Wedge(0,end_r, start_t, end_t)
        patches.append(wedge)
```

```python
collection = PatchCollection(patches)
```
